```python
x  = np.arange(0,10,0.01)
```

## $\delta$-Function Lineshape
The Diract $\delta$ function has the following form

$$ y = \begin{cases}
        1 & \text{if } x=x_0 \\
        0 & \text{otherwise}
      \end{cases} $$
      
This means that this curve depends on the $x_0$ (or `x0`) varibale.

### Exercise 1
Play around with the `x0` value in box below to see how the graph changes. Remember to press [shift][enter] while the box/cell is highlighted to run the code. **Answer question 1a afterwards**.

```python
x0 = 5

y = np.array([1 if (i-x0)==0 else 0 for i in x])
plt.plot(x, y)
plt.show()
```

## Lorentzian Lineshape

The Lorentzian lineshape function is
$$y = \frac{\Gamma}{\pi \left( \Gamma^2 + (x - x_0)^2 \right)}$$
This means that we need two variables to determine the graph: $\Gamma$ (or `gamma`) and $x_0$ (or `x0`).
### Exercise 2
Try changing the values of these two variables below to see how these affect the graph, then **answer questions 2a and 2b**.

```python
gamma = 0.8
x0 = 5
y = gamma / (np.pi * (gamma**2 + (x-x0)**2)) # this is the Lorentzian function from the equation above
plt.plot(x,y)
plt.show()
```

**Question 2a:** How does `gamma` affect the Lorentzian curve?

**Question 2b:** How does `x0` affect the Lorentzian curve?

## Gaussian Lineshape
A Gaussian lineshape function is
$$y = \frac{1}{\sqrt{\pi}} e^{-(x-x_0)^2/\Gamma^2}$$
Once again, this function depends on two variables: `gamma` and `x0`. 
### Exercise 3
Try changing the values of these two variables below to see how these affect the graph, then **answer questions 3a and 3b**.

```python
gamma = 0.8
x0 = 5
y = (1/np.sqrt(np.pi)) * np.exp(-1*(x-x0)**2/gamma**2) # This is the Gaussian lineshape function from above
plt.plot(x,y)
plt.show()
```

**Question 3a:** How does `gamma` affect the Gaussian curve?

**Question 3b:** How does `x0` affect the Gaussian curve?

```python
gamma_lorentzian = 0.5
x0_lorentzian    = 5
y_lorentzian     = gamma_lorentzian / (np.pi * (gamma_lorentzian**2 + (x-x0_lorentzian)**2))

gamma_gaussian = 0.5
x0_gaussian = 5
y_gaussian = (1/np.sqrt(np.pi)) * np.exp(-1*(x-x0_gaussian)**2/gamma_gaussian**2)

plt.plot(x, y_lorentzian, label='Lorentzian')
plt.plot(x, y_gaussian, label='Gaussian')
plt.legend()
plt.show()
```
